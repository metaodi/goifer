```python
import goifer
import pandas as pd
```

```python
client = goifer.Client("canton_zurich")
```

```python
# check fields available for search
client.schema("Mitglieder")['annotation_documentation']
```

```python
factions = client.search("Behoerden", query="GremiumTyp adj Fraktion")
df_fac = pd.DataFrame(factions)
df_fac
```

```python
member_dfs = []
for faction in factions:
    members = client.search(
        "Mitglieder",
        query=f"dauer_end >= \"2020-11-19 00:00:00\" and dauer_start <= \"2020-11-19 00:00:00\" and gremium adj {faction['kurzname']} sortBy name/sort.ascending vorname/sort.ascending",
    )
    norm_member = pd.json_normalize(members)
    df = pd.DataFrame(norm_member)
    member_dfs.append(df)
df_member = pd.concat(member_dfs).reset_index(drop=True)
df_member
```
